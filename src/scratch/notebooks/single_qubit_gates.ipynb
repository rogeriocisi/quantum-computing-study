# 🌌 Single Qubit Gates Exploration

This notebook demonstrates the behavior of fundamental quantum gates (H, X, Y, Z, S, T) on a single qubit, visualizing the statevectors using Dirac notation.

```python
# 1. Environment Setup
import sys
import os
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Add project root to Python execution path
sys.path.append(os.path.abspath(os.path.join('..')))
```

## Initial State $|0\rangle$

```python
qc = QuantumCircuit(1)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Hadamard Gate ($H$)
Creates a perfect superposition: $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$.

```python
qc = QuantumCircuit(1)
qc.h(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Pauli-X Gate ($X$)
The quantum NOT gate. Flips $|0\rangle$ to $|1\rangle$.

```python
qc = QuantumCircuit(1)
qc.x(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Pauli-Y Gate ($Y$)
Rotates $\pi$ radians around the Y-axis.

```python
qc = QuantumCircuit(1)
qc.y(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Pauli-Z Gate ($Z$)
Flips the phase of the $|1\rangle$ state.

```python
qc = QuantumCircuit(1)
qc.z(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Phase Gate ($S$)
Applies a $\pi/2$ phase shift.

```python
qc = QuantumCircuit(1)
qc.s(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## $\pi/4$ Phase Gate ($T$)
Applies a $\pi/4$ phase shift.

```python
qc = QuantumCircuit(1)
qc.t(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

## Gates on Initial State $|1\rangle$
First, we prepare the qubit in state $|1\rangle$ using an $X$ gate.

```python
qc = QuantumCircuit(1)
qc.x(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### Hadamard Gate ($H$) on $|1\rangle$
Transforms $|1\rangle$ to $|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### Pauli-X Gate ($X$) on $|1\rangle$
Flips $|1\rangle$ back to $|0\rangle$.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.x(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### Pauli-Y Gate ($Y$) on $|1\rangle$
Rotates $\pi$ radians around the Y-axis.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.y(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### Pauli-Z Gate ($Z$) on $|1\rangle$
Flips the phase of the $|1\rangle$ state: $Z|1\rangle = -|1\rangle$.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.z(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### Phase Gate ($S$) on $|1\rangle$
Applies a $\pi/2$ phase shift: $S|1\rangle = i|1\rangle$.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.s(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```

### $\pi/4$ Phase Gate ($T$) on $|1\rangle$
Applies a $\pi/4$ phase shift: $T|1\rangle = e^{i\pi/4}|1\rangle$.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.t(0)
psi = Statevector.from_instruction(qc)
psi.draw('latex')
```
